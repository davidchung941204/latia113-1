```python
def wtf(num):
  return num + 1

test_values = [0.0, 0.0011, 0.25, 0.5,0,54, 0.75, 1.0]
outcome = [wtf(num) for num in test_values]

print("test value: ",test_values)
print("outcome: ",outcome)
```

```python
def a(num):
  if (num == 0)or(num == 1):
    return 1
  else:
    return(num+a(num-1))
a(5)
```
